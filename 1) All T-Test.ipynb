# T-Test

A t-test is a type of inferential statistic which is used to determine if there is a significant difference between the means of two groups which may be related in certain features

T-test has 2 types : 

#### 1. one sampled t-test 

#### 2. two-sampled t-test.

### One Sample T-test: (stats.ttest_1samp)

- This test help us to understand that is their any difference between population mean and sample.

Here we are comparing population and sample

Validation: 

Is there any difference between sample mean and population mean

* H0 = There is no difference between the mean

* H1 = There is difference between the mean



Use One Sample T-test to compute the p-value

* if p-value < 0.05: REJECT H0 & ACCEPT H1

* if p-value > 0.05: REJECT H1 & ACCEPT H0


Than validing something based on the evidences

```python
ages=[10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70]
```

```python
import pandas as pd
import numpy as np

# Population mean

age_mean = np.mean(ages)
print(age_mean)
```

```python
# Take Sample

sample_size = 10
age_sample = np.random.choice(ages, sample_size)

age_sample
```

```python
# Calculating that the sample mean and population mean is having difference or not

from scipy.stats import ttest_1samp

ttest, pvalue = ttest_1samp(age_sample, 30) # 30 = len(ages)

print(f"Stats T-test value: {ttest} & P-value: {pvalue}")
```

```python
if pvalue < 0.05:
    print("REJECT H0 & ACCEPT H1")
else:
    print("REJECT H1 & ACCEPT H0")
```

### Two Sample T-test: (stats.ttest_ind)


Here we are comparing 2 different groups

Validation: 

Is there any difference between group1 and group2

* H0 = There is no difference

* H1 = There is difference



Use One Sample T-test to compute the p-value

* if p-value < 0.05: REJECT H0 & ACCEPT H1

* if p-value > 0.05: REJECT H1 & ACCEPT H0


Than validing something based on the evidences

```python
import scipy.stats as stats
import pandas as pd
import numpy as np
import math


np.random.seed(6)

school_ages = stats.poisson.rvs(loc=18, mu=35, size=1500)      # Age=18, Age_Mean=35, Population=1500
classA_ages = stats.poisson.rvs(loc=18, mu=30, size=60)

np.random.seed(12)
classB_ages = stats.poisson.rvs(loc=18, mu=33, size=60)
```

```python
print(f"Population Mean: {school_ages.mean()}\nClassA_ages Mean: {classA_ages.mean()}\nClassB_ages Mean: {classB_ages.mean()}")
```

```python
ttest, pvalue = stats.ttest_ind(a=classA_ages, b=classB_ages, equal_var=False) # group1, group2

print(f"Stats T-test value: {ttest} & P-value: {pvalue}")
```

```python
if pvalue < 0.05:
    print("REJECT H0 & ACCEPT H1")
else:
    print("REJECT H1 & ACCEPT H0")
```

### Paired T-test:  (stats.ttest_rel)

- Used when you want to check how different, samples from the same group are, you can go for a paired T-test

```python
weight1=[25,30,28,35,28,34,26,29,30,26,28,32,31,30,45]
weight2=weight1+stats.norm.rvs(scale=5,loc=-1.25,size=15)
```

```python
print(weight1)
print('\n')
print(weight2)
```

```python
weight_df = pd.DataFrame({"Weight_10":np.array(weight1), 
                          "Weight_20":np.array(weight2), 
                          "Weight_Changed": np.array(weight2) - np.array(weight1)})

weight_df
```

```python
ttest, pvalue = stats.ttest_rel(a=weight1, b=weight2)

print(f"Stats T-test value: {ttest} & P-value: {pvalue}")
```

```python
if pvalue < 0.05:
    print("REJECT H0 & ACCEPT H1")
else:
    print("REJECT H1 & ACCEPT H0")
```
